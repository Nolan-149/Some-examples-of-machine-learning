## 了解神经网络的起源与发展
### 神经元简介
> + 19世纪末，科学家在生物、生理学领域创建了神经元学说。人工神经元的研究起源于脑神经元学说，人们认识到复杂的神经网络系统是由数据繁多的神经元组成的。神经元是大脑中相互链接的神经细胞，它可以处理和传递化学信号和电信号。
> + 人类的大脑皮层中包含百亿个的神经元，平均每立方毫米就由数万个，他们相互联结形成神经网络。人类通过神经网络来实现机体与内外环境的联系，协调全身的各种机能活动。神经元细胞的形态比较特殊，具有许多突起，分为细胞体、轴突和树突这3个部分。
> + <img src="./imgs/chap11/fig11_01.png" width=50%>
> + https://www.bilibili.com/video/BV1Fu411D7an/?spm_id_from=333.788.recommend_more_video.9&vd_source=9322a2ac8c0fc2033c5ead7208976489
> + 神经元中具备和计算机中的“1”和“0”原理几乎完全一样的两种常规工作状态：
>> + 兴奋（1）   
>> + 抑制（0）
>> + 当传入的神经冲动使细胞膜电位升高且超过阈值时，细胞进入兴奋状态，产生神经冲动并由轴突输出；
>> + 当传入的神经冲动使细胞膜电位下降且低于阈值时，细胞进入抑制状态，便没有神经冲动输出。

### 神经网络的起源
> + “神经网络”最早来源于生物学，而本书所指的神经网络实际是应用在人工智能领域的。最早在1943年，美国精神解剖学家沃伦·麦克洛奇和数学家沃尔特·皮茨就提出了神经元的数学描述和结构，及M-P模型，并证明了只要有足够多简单的神经元，在这些神经元互相连接并同步运行的情况下，模型可以模拟任何计算函数。这项开创性的工作，被认为是人工神经网络研究方向的划时代的起点和基础。

## 掌握神经网络的原理
### 神经网络的原理
> + 神经网络是一种算法结构，由大量的互连的人工神经元组成，它们按顺序堆叠在一起，以“层”的形式形成数百万个连接，一般情况下“层”之间仅通过输入和输出与他们之间的神经元层互连，让机器能够进行自我学习。

## 掌握神经网络中的激活函数
### 了解激活函数
> + 1.什么是激活函数
>> + 激活函数是神经网络重要组成部分之一，神经网络在不使用激活函数的情况下，如果每一层的输入都是上一层的线性输出，那么无论神经网络有多少层，输出与输入都是线性相关的，有没有隐藏层没有太大区别。
>> + <img src="./imgs/chap11/fig11_12.png" width=50%>
>> + 引入非线性函数即激活函数，这样多层神经网络的输出不再是输入的线性组合，可以逼近任意函数，多层神经网络的真正效果才能显现出来。
>> + <img src="./imgs/chap11/fig11_13.png" width=50%>
 
> + 2.常用的激活函数
>> + Identity函数
>> + Sigmoid函数
>> + tanh函数
>> + relu函数
>> + leakyrelu函数
>> + cos函数
>> + sinc函数

> + relu函数
>> + Rectified Linear Unit，是一种人工神经网络中常用的激活函数，通常意义下，其指代数学中的斜坡函数，即
$$
f(x)=\max (0, x)
$$
在神经网络中，ReLU函数作为神经元的激活函数一，为神经元在线性变换$\mathbf{w}^{\mathrm{T}} \mathbf{x}+\mathrm{b}$之后的非线性输出结果。换言之，对于进入神经元的来自上一层神经网络的输入向量x，使用ReLU函数的神经元会输出:
$$
\max \left(0, \mathbf{w}^{\mathrm{T}} \mathbf{x}+\mathrm{b}\right)
$$
至下一层神经元或作为整个神经网络的输出（取决现神经元在网络结构中所处位置）。
>> + 相比于传统的神经网络激活函数，诸如逻辑函数（Logistic sigmoid）和tanh等双曲函数，ReLU函数有着以下几方面的优势：
>>> + 仿生物学原理：相关大脑方面的研究表明生物神经元的讯息编码通常是比较分散及稀疏的。通常情况下，大脑中在同一时间大概只有1%-4%的神经元处于活跃状态。使用线性修正以及正规化;（regularization）可以对机器神经网络中神经元的活跃度（即输出为正值）进行调试；相比之下，逻辑函数在输入为0时达到 0.5，即已经是半饱和的稳定状态，不够符合实际生物学对模拟神经网络的期望。不过需要指出的是，一般情况下，在一个使用ReLU的神经网络中大概有50%的神经元处于激活态。
>>> + 更加有效率的梯度下降以及反向传播：避免了梯度爆炸和梯度消失问题;
>>> + 简化计算过程：没有了其他复杂激活函数中诸如指数函数的影响；同时活跃度的分散性使得神经网络整体计算成本下降.

## 掌握神经网络中的参数调节

### 分析数据集并展现
> + 神经网络模型对激活函数的使用是一个很重要的非线性矫正过程，为了让激活函数发挥更大的作用，就需要不断地进行参数调节。
> + 下面通过利用鸢尾花数据集对神经网络模型进行训练，通过参数调节体会各个参数对模型的影响。

> + 【示例】生成数据集并了解重要参数

```python
#导入MLP神经网络
from sklearn.neural_network import MLPClassifier
#载入pandas
import pandas as pd
#导入鸢尾花数据集
from sklearn.datasets import load_iris
#导入数据集拆分工具
from sklearn.model_selection import train_test_split
#生成记录数为300，三个分类的鸢尾花数据集
iris = load_iris()  # 返回datasets.base.Bunch类型(字典格式)
X = iris.data[:,:2]
y = iris.target
#将数据集拆分为训练数据集和测试数据集
X_train, X_test, y_train, y_test = train_test_split(X,y,random_state=42)
#定义MLP分类器
mlp = MLPClassifier(solver='lbfgs')
mlp.fit(X_train, y_train)
```

**重要参数**
> * `activation` 非线性激活函数，包含`identity`、`logistic`、`tanh`、`relu`，默认`relu`
> * `alpha`  和线性模型的alpha值一样，是L2惩罚项，用来控制正则化程度，默认0.0001
> * `hidden_layer_sizes` 表示隐藏层的数量，默认是`[100,]`，1个隐藏层，其节点数是100;
若改成`[10,10]`，则表示模型中有两个隐藏层，每层有10个节点
> * `solver` 该参数决定了逻辑回归损失函数的优化方法，包含`liblinear`、`Ibfgs`、`newton-cg`、`sag`等!
>> * `liblinear`:       使用开源的liblinear库实现，内部使用坐标轴下降法来迭代优化损失函数。
>> * `Ibfgs`:            牛顿法的一种，利用损失函数二阶导数矩阵来迭代优化函数。
>> * `newton-cg`:   也是牛顿法的一种，利用损失函数二阶导数矩阵来迭代优化函数。
>> * `sag` ：              随机平均梯度下降，是梯度下降的变种。

> * 【示例】多层感知机分类可视化展现
>> * 知识点
>>> * np.c_和np.r_用于连接两个矩阵
>>> * np.r_中的r是row的缩写：是按行叠加
>>> * np.c_中的c是column的缩写：是按列叠加

```python
#导入画图工具
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap

#使用不同色块表示不同分类
cmap_light = ListedColormap(['#FFAFAA', '#AAFFDA', '#FAAAFF'])
cmap_bold = ListedColormap(['#FF00AA', '#00FF0F', '#0FA0FF'])
x_min, x_max = X_train[:, 0].min() - 1, X_train[:, 0].max() + 1
y_min, y_max = X_train[:, 1].min() - 1, X_train[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, .02),
                     np.arange(y_min, y_max, .02))
Z = mlp.predict(np.c_[xx.ravel(), yy.ravel()])

Z = Z.reshape(xx.shape)
plt.figure()
plt.pcolormesh(xx, yy, Z, cmap=cmap_light)

#将数据特征用散点图表示出来
plt.scatter(X[:, 0], X[:, 1], c=y, edgecolor='k', s=60)
plt.xlim(xx.min(), xx.max())
plt.ylim(yy.min(), yy.max())
#设定图题
plt.title("MLPClassifier:solver=lbfgs")
#显示图形
plt.show()
```

```python
print('MLPClassifier:solver=lbfgs 得分：{:.2f}%'.format(mlp.score(X_train,y_train)*100))
```

### 参数调节
> + 调整隐藏层节点数:      `hidden_layer_sizes`  
> + 调节隐藏层数量:        ` hidden_layer_sizes` 
> + 调节激活函数类型:      `activation` 
> + 调整模型复杂度:         `alpha`   


> * 【示例】调整隐藏层节点数（1个隐藏层20个节点）

```python
#设定隐藏层中的节点数为20
mlp_20=MLPClassifier(solver='lbfgs', hidden_layer_sizes=[20])
mlp_20.fit(X_train, y_train)
Z1 = mlp_20.predict(np.c_[xx.ravel(), yy.ravel()])

Z1 = Z1.reshape(xx.shape)
plt.figure()
plt.pcolormesh(xx, yy, Z1, cmap=cmap_light)

#使用散点图画出X
plt.scatter(X[:, 0], X[:, 1], c=y, edgecolor='k', s=60)
plt.xlim(xx.min(), xx.max())
plt.ylim(yy.min(), yy.max())
#设置图题
plt.title("MLPClassifier:nodes=20")
#显示图形
plt.show()
```

```python
print('MLPClassifier:hidden_layer_sizes=[20] 得分：{:.2f}%'.format(mlp_20.score(X_train,y_train)*100))
```

> * 【示例】调整隐藏层节点数（2个隐藏层,每层20个节点）

```python
#设置神经网络有两个节点数为20的隐藏层
mlp_2L=MLPClassifier(solver='lbfgs', hidden_layer_sizes=[20,20])
mlp_2L.fit(X_train, y_train)
Z1 = mlp_2L.predict(np.c_[xx.ravel(), yy.ravel()])
#用不同色彩区分分类
Z1 = Z1.reshape(xx.shape)
plt.figure()
plt.pcolormesh(xx, yy, Z1, cmap=cmap_light)
#用散点图画出X
plt.scatter(X[:, 0], X[:, 1], c=y, edgecolor='k', s=60)
plt.xlim(xx.min(), xx.max())
plt.ylim(yy.min(), yy.max())
#设定图题
plt.title("MLPClassifier:2layers")
#显示图形
plt.show()
```

```python
print('MLPClassifier:hidden_layer_sizes=[20,20] 得分：{:.2f}%'.format(mlp_2L.score(X_train,y_train)*100))
```

 > * ↑ 由以上对比结果图可以看出，隐藏层数增加，分类决定边界越平滑，但此处模型的评分反而下降。
  
  
 > * ↓【示例】调整激活函数(激活函数默认relu，现修改为tanh)

```python
#设置激活函数为tanh
mlp_tanh=MLPClassifier(solver='lbfgs', hidden_layer_sizes=[20,20],
                    activation='tanh')
mlp_tanh.fit(X_train, y_train)
#重新画图
Z2 = mlp_tanh.predict(np.c_[xx.ravel(), yy.ravel()])

Z2 = Z2.reshape(xx.shape)
plt.figure()
plt.pcolormesh(xx, yy, Z2, cmap=cmap_light)
#散点图画出X
plt.scatter(X[:, 0], X[:, 1], c=y, edgecolor='k', s=60)
plt.xlim(xx.min(), xx.max())
plt.ylim(yy.min(), yy.max())
#设置图题
plt.title("MLPClassifier:2layers with tanh")
#显示图形
plt.show()
```

```python
print("MLPClassifier:activation='tanh'得分：{:.2f}%".format(mlp_tanh.score(X_train,y_train)*100))
```

  > * ↑ tanh激活函数比relu激活函数下的模型更平滑，模型的评分也有显著增长。
 
 > * ↓ 【示例】调整模型复杂度控制
>> * alpha越大惩罚力度越大，控制模型复杂度,默认0.0001，现修改为1

```python
#修改模型的alpha参数
mlp_alpha=MLPClassifier(solver='lbfgs', hidden_layer_sizes=[20,20],
                    activation='tanh',alpha=1)
mlp_alpha.fit(X_train, y_train)
#重新绘制图形
Z3 = mlp_alpha.predict(np.c_[xx.ravel(), yy.ravel()])

Z3 = Z3.reshape(xx.shape)
plt.figure()
plt.pcolormesh(xx, yy, Z3, cmap=cmap_light)
#散点图画出X
plt.scatter(X[:, 0], X[:, 1], c=y, edgecolor='k', s=60)
plt.xlim(xx.min(), xx.max())
plt.ylim(yy.min(), yy.max())
#设定图题
plt.title("MLPClassifier:alpha =1")
#显示图形
plt.show()
```

```python
print("MLPClassifier:activation='tanh',alpha=1 得分：{:.2f}%".format(mlp_alpha.score(X_train,y_train)*100))
```

 > * ↑增大L2惩罚项力度alpha，使得模型更简单，模型的评分有所增长。

**关于MLP模型参数调整小结**
> * 对于初学者来说，设置参数可以参考如下原则：
>> * 神经网络中隐藏层的节点数约等于训练集的特征数量，但一般不要超过500。
>> * 在开始训练模型的时候，可以让模型尽量复杂，然后再通过调节alpha值来调节模型的复杂度。
