# K-Means聚类

> <img src="./imgs/chap05/fig05_01.png" width="40%">

## 模型构建与训练

```python
# 导入KMeans
from sklearn.cluster import KMeans
# 模型构建&训练,设置n_clusters=3
model = KMeans(n_clusters=3 ,random_state=77).fit(auto_scaled)    #为使复现效果与PPT效果相同，设置随机种子为77
```

## 聚类结果分析

```python
# 聚类结果
## (1) Clusters的标签
auto_label = model.labels_ 
## (2)Clusters的聚类中心
auto_cluster = model.cluster_centers_ 
# 输出聚类中心
print("聚类中心:\n",auto_cluster)
# 查看类标签
auto_label
```

## 模型优化
> + `KMeans(n_clusters=i)`需要人为确定，是个超参

```python
# 选择不同k值比较聚类效果
print("选择不同k值比较聚类效果")
for i in [2,3,4,6]:
    # 模型构建&拟合
    model = KMeans(n_clusters=i,random_state=77,).fit(auto_scaled)   #为使复现效果与PPT效果相同，设置随机种子为77
    # 获取标签
    labels = model.labels_
    # 模型评价：计算模型的轮廓系数
    silh_score = metrics.silhouette_score(auto_scaled, labels)
    print("K={} 时 轮廓系数(Silhouette Coefficient):{:.4f}".format(i,silh_score))
```

**↑♎分析：**从结果看，如果将数据聚成两个簇，其效果更好些。

# DBSCAN聚类
> DBSCAN(Density-Based Spatial Clustering of Applications with Noise):
>> 是一个比较有代表性的**基于密度的聚类算法**。与划分和层次聚类方法不同，它将簇定义为密度相连的点的最大集合，能够把具有足够高密度的区域划分为簇，并可在噪声的空间数据库中发现任意形状的聚类。

> DBSAN思想：
>> ●DBSCAN主要思想是把样本空间中不同的高密度区域划分为不同簇 <br>
●对于一个样本点，如果在它的半径为$𝜀$的超球体(邻域)内的样本点个数大于$𝑁_{𝑚𝑖𝑛}$，则认为这个样本点在一个高密度区域内，称这个点为核心对象 <br>
●对于两个都处在高密度区域的样本点，如果他们之间的距离比较接近（即互相在对方的邻域内），则认为它们同属于一个簇 <br>

<img src="./imgs/chap05/fig05_04.png" width="40%">

>重点区分： 
>> 1.直接密度可达(directly density-reachable),邻域内的点，属于本簇，高密度区域。   <br>
2.密度可达(density-reachable)，邻域边界上的点，也属于本簇，但不属于高密度点；<br>
3.邻域圈不住的点，孤立点，不属于本簇。

> 算法流程：
>> 1.从一个点开始，对其半径为$𝜀$的邻域内的点赋予同一个簇标记，并寻找邻域内的其它核心对象;<br>
2.对于找到的核心对象，重复步骤1;<br>
3.重复上述过程直到无法找到新的高密度点为止，认为一个簇被完整找出;<br>
4.从未被访问过的点开始，寻找下一个簇，最终没有出现在任何簇内的点被标记为孤立点.<br>

<img src="./imgs/chap05/fig05_05.gif" width="50%">

> 重要参数;
>> |参数|	功能|
|---:|---:|
|eps	|邻域的半径，默认为0.5|
|min_samples	|确定核心对象的邻域内的最小样本数，默认为5|
|metric	|点之间的距离度量方式，默认为“euclidean”（欧式距离）|

# 小结
> |算法	|优点	|缺点|
|---:|---:|---:|
|K-Means	|实现简单，直观，支持多种距离计算方式	|需要指定簇数`n_clusters`，且效果依赖于初始质心的选择，容易陷入局部最优。同时不易处理非‘球’状数据，易受离群值影响|
|MeanShift	|不需要指定簇的个数，可以处理非‘球’状数据	|效果依赖于带宽的选择|
|层次聚类	|能产生不同层级的聚类结果，更加灵活	|需要指定`n_clusters`,空间和时间复杂度要高于K-Means聚类|
|DBSCAN	|不需要指定簇的个数，可以找出任何形状的簇，能有效分辨噪音样本	|对参数`eps`和`min⁡_samples`的设置很敏感|
|谱聚类|还需进一步研究：实现图分割|需要指定簇的个数,`cluster.SpectralClustering`|
